```python
# only get the data you need - which is basically the whole point of an API to begin with
# https://api-v3.mbta.com/routes/?fields[route]=short_name,long_name

long_names = []
route_ids = []
# names_dict = {}

def get_route_names():
    response_routes = requests.get("https://api-v3.mbta.com/routes/?fields[route]=short_name,long_name,id&filter[type]=0,1").json()
    data_routes = response_routes["data"]
    for route in data_routes:
        # print(route)
        # print(route['attributes']['long_name'])
        long_name = route['attributes']['long_name']
        route_id = route['id']
        long_names.append(long_name)
        route_ids.append(route_id)
        # names_dict[route_id] = long_name
        # print(long_name)
        # print(route_id)
    name_tuple = zip(long_names, route_ids)
    return name_tuple
        
        
get_route_names()
print(tuple(get_route_names()))
route_names_ids = tuple(get_route_names())
print("long names: " + str(long_names))
print("route ids" + str(route_ids))
#print(names_dict)
```

```python
## MAYBE LOOP OVER THE orig get_stops func instead...

## Remember that route == route_id

def get_stops(route_id):
    endpoint_stops = "https://api-v3.mbta.com/stops"
    stop_q_params = {"fields[stop]": "name", "include": "route", "filter[route]": route_id}
    response_stops = requests.get(endpoint_stops, stop_q_params).json()
    stop_data = response_stops["data"]
    route_stops = []
    for stop in stop_data:
        stop_name = stop["attributes"]["name"]
        route_stops.append(stop_name)
        # print(stop_name)
    return route_stops

# get_stops("Red")
print("from get_stops - TESTING")
print(get_stops("Red"))
print("-----LINE BREAK-----")


route_dict = {}
def make_route_stop_dict():    
    for route_id in route_ids:
        # get_stops(route_id)
        # print("{route_id} stops are {route_stops}".format(route_id = route_id, route_stops = get_stops(route_id)))
        route_dict[route_id] = get_stops(route_id)
        # print("this is the route_dict" + route_dict)
    return route_dict

# make_route_stop_dict()
print(make_route_stop_dict())
print(route_dict)
```

## Now let's put it together

```python
class Route():
    def __init__(self, long_name, route_id, stops):
        self.long_name = long_name
        self.route_id = route_id
        self.stops = stops
        #not sure this goes here
        num_stops = len(stops)
        
    def __repr__(self):
        return str(self)
```

```python
#list of class objects
routeList = []
for long_name, route_id in route_names_ids:
    routeList.append(Route(long_name, route_id, route_dict[route_id]))

for route in routeList:
    print('long_name: {}, route_id {}, stops {}'.format(route.long_name, route.route_id, route.stops))
        
```
